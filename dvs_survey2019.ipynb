# An exploratory analysis of the 2019 Data Visualization Society survey results

### Get an overview of the survey questions and data

Full copy of the survey questions and response options are available [here](https://docs.google.com/forms/d/e/1FAIpQLSe-CMHrt1gYOxO9Tdk6cjyLIpJnAJMDjS94OBDHi6YU2g3sBA/viewform?fbzx=3061750194003575665).

```python
import pandas as pd

results = pd.read_csv('https://raw.githubusercontent.com/data-visualization-society/data_visualization_survey/master/data/cleaned_survey_results_2019.csv')

# Quick recap of survey questions. 
results.columns
```

### Brainstorm possible questions

* How does the degree of focus on dataviz in practitioners' jobs affect the number of hours spent on creating dataviz? In other words, how many hours can you expect to spend on direct dataviz work when it's your core focus, secondary focus or just one element?
* How does the experience or desire of spending time on dataviz vary across role types (e.g. analyst vs developer)? 
* Is there a difference in the tools that practitioners use when they get to choose the tools versus when the choices are made for them? Does this vary across role types?
* What could improve the practice of dataviz - frustration, what others don't get, one thing to change
* How is dataviz normally used in the organisation and might this be tied to certain frustration?

### Prepare data for answering

```python
degree_of_focus = ['Data visualization is the focus of my job', 'Data visualization is an important secondary part of my job, but not the focus', 'Data visualization is one of several other things I do in my job, but not a primary or secondary part of my role']
for i in degree_of_focus:
    print(i, results.loc[results['What focus is data visualization in your work?'] == i, 'Hours a week focused on creating/implementing/productizing data visualizations?'].describe())
```

### Try out some visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style= "whitegrid", palette = 'GnBu_r')
sns.swarmplot(order = degree_of_focus, y = results['What focus is data visualization in your work?'], x = results['Hours a week focused on creating/implementing/productizing data visualizations?'])
plt.show()
```
